```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ravitejakotharu/salary-datacsv")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(path + "/salary_data.csv")
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, :1], data.iloc[:, 1:2], test_size=0.2, random_state = 42)
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY VS EXPERIENCE (training set)')
plt.xlabel('Experience in Years' )
plt.ylabel('Salary in Rupees')
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY VS EXPERIENCE (test set)')
plt.xlabel('Experience in Years' )
plt.ylabel('Salary in Rupees')
plt.show()
```

## Predicting fare of a taxi ride in chicago


```python
import kagglehub

path = kagglehub.dataset_download("adelanseur/taxi-trips-chicago-2024")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(path + "/Taxi_Trips_-_2024_20240408.csv")
```

```python
data = data[['Trip Miles', "Fare"]]
```

```python
data.dropna(inplace=True)
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, :1], data.iloc[:, 1:2], test_size=0.2, random_state = 42)
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Distance vs Fare (training set)')
plt.xlabel('Distance' )
plt.ylabel('Fare')
plt.show()
```
