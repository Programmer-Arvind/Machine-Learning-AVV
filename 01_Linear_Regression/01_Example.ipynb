```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ravitejakotharu/salary-datacsv")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
data = pd.read_csv(path + "/salary_data.csv")
```

```python
# Splitting the data into 80 20 split
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, :1], data.iloc[:, 1:2], test_size=0.2, random_state = 42)
```

```python
# Fitting the model
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY VS EXPERIENCE (training set)')
plt.xlabel('Experience in Years' )
plt.ylabel('Salary in Rupees')
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title("SALARY VS EXPERIENCE (testing set)")
plt.xlabel("Experience in years")
plt.ylabel("Salary in Rupees")
plt.show()
```
